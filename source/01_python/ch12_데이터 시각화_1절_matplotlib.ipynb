### 1) subplot() 함수로 서브플롯 추가

```python
import numpy as np
x = np.arange(0,10,0.01)
plt.figure(figsize=(8,6))
# subplot(전체화면 가로개수, 전체화면 세로개수, 위치 할당)
plt.subplot(2,1,1) ; plt.plot(x, np.sin(x))
plt.subplot(2,2,3) ; plt.plot(x, np.cos(x))
plt.subplot(2,2,4) ; plt.plot(x, np.sin(x)*np.cos(x)) # plt.subplot(224) 와 동일
plt.show()
```

### 3) bar(), barh(), axvline(수직선), axhline(수평선)

```python
x = [1,2,3]
y = [3,5,2]
fig, axes = plt.subplots(ncols = 3, figsize = (15,3))
axes[0].bar(x,y, color='limegreen',)
axes[1].barh(x,y, color = 'r') # x,y 좌표를 y,x로 사용하므로 plot을 추가할 때 x,y => y,x로 대입
axes[0].plot(x,y, 'ko-.')
axes[1].plot(y,x, 'ko-', linewidth = 2)
axes[2].axvline(0.5,color = 'g') # x=0.5 자리에 수직선 추가
axes[2].axhline(0.5,color = 'r', linestyle = '-.') # x=0.5 자리에 수직선 추가
```

### 4) hist(도수분포표;histogram)

```python
fig, axes = plt.subplots(nrows =2, ncols = 2, figsize = (10,6))
axes[0,0].hist(iris.sepal_length, bins = 10, color = 'r') # 히스토그램 ; 도수분포표
axes[0,1].hist(iris.sepal_length, bins = 15, cumulative = True) # 누적 히스토그램 ; 누적 도수분포표 => 총개수에 수렴 ★
axes[1,0].hist(iris.sepal_length, bins = 5, orientation = 'horizontal') # 수평 히스토그램
axes[1,1].hist(iris.sepal_length, bins = 10, histtype = 'step')
plt.show()
```

### 4) 축과 눈금
- https://matplotlib.org/3.1.0/api/axis_api.html (교안 33page)

```python
x = np.linspace(0,10,100)
y = np.cos(x)

fig, axes = plt.subplots(nrows=2, ncols=2 , figsize = (16,4))
# 서브플롯 끼리의 여백 조정 : hspace : 위아래 여백, wspace : 좌우 여백
plt.subplots_adjust(hspace=0.5, wspace=0.2) # 간격
plt.suptitle('메인 그래프', fontsize = 20, fontproperties = font_prop)

# 0행 0열 그래프 : set()을 통해 만들기
axes[0,0].scatter(x,y, marker='.',s = 20, c = 'limegreen')
axes[0,0].set(title='기본값',
              ylabel = 'y값',
              xlabel ='x값',
              xlim=(-1,11), # x 눈금 범위
              ylim=(-2,2), # y 눈금 범위
              xticks = [0,4,6,8,12], # x축 눈금
              xticklabels=['원점', '4점', 6 ,'8점', '최대'],
              yticks = [-2,-1,0,1,2],
              yticklabels = ['limit_min', 'real_min' , '0', 'real_max', 'limit_max']
             ) 
        

# 0행 1열 그래프 : 함수형으로 만들기
axes[0,1].scatter(x,y, marker='^', color = 'r',s = 10)
axes[0,1].set_title('함수형으로 표현하면 세부지정 가능', fontsize = 15, fontproperties = font_prop)
axes[0,1].set_xlim(0,5) # set(xlim=[0,5])
axes[0,1].set_ylim(-2,2)

# 1행 0열
axes[1,0].scatter(x,y, c= 'b',s = 0.5)
axes[1,0].spines['top'].set_visible(False) # 축 경계 지우기

# 1행 1열
axes[1,1].scatter(x,y, c = 'c', s = 0.5)
axes[1,1].spines['bottom'].set_position(('outward',10))
axes[1,1].spines['top'].set_position(('outward',10))
axes[1,1].grid(True)

plt.show()
```

### 5) 축 공유

```python
x = np.arange(0,10) 
y1 = 0.5 * x ** 2
y2 = -1*y1

fig, ax1 = plt.subplots(1,1 ,figsize = (10,3))
ax1.plot(x,y1, 'g^:')
ax1.set(title='축공유', xlabel ='Xdata')
ax1.set_ylabel('Y1 data', color = 'g')

ax2 = ax1.twinx() # ax1과 x축을 공유하는 새로운 ax2 => 새로운 y축 생성
ax2.plot(x,y2,'bv--')
ax2.set_ylabel('Y2 data', color = 'b')

ax3 = ax1.twiny() # ax1과 y축을 공유하는 ax3 => 새로운 x축 생성
ax3.plot(-x,y1, 'ro-.')
ax3.set_xlabel('-X data', color = 'r')

# ax4 = ax1.twinx()
# ax4.plot(x,y1**2, 'k') # => 그래프 오염(y2축이 너무 난잡해짐)
plt.show()
```
