## 9.1 Groupby 그리고 pivot_table
- cf) pivot_table
- Groupby는 통계처리에 용이하다

### groupby

````python
# 1) 그룹화 방법 1
#    df.groupby(by = [그룹핑할열]).기초통계함수()[[함수를 적용해서 출력할 열들]]
#    df.groupby(by = [그룹핑할열])[[함수를 적용해서 출력할 열들]].기초통계함수()
#                               [[함수를 적용해서 출력할 열들]] 생략 시 그룹핑할 열을 제외한 모든 열 적용
iris_grouped = iris.groupby(by='species')
iris_grouped = iris.groupby(by=iris.species)
iris_grouped = iris.groupby(by=[iris.species])
for group, data in iris_grouped :
    print(group)
    display(data.sample(2).sort_index())
````

````python
iris.groupby(by='species').mean()[['petal_length','petal_width']]
iris.groupby(by='species')[['petal_length','petal_width']].mean() # 문법
````

````python
iris.groupby(by='species').mean()
````

````python
iris_grouped_sum = iris.groupby(iris.species).median()
iris_grouped_sum # 특정 행을 뽑기에 용이
iris_grouped_sum.loc[['setosa','versicolor']]
````

````python
iris.groupby(iris.species, as_index= False).median() # as_index = False를 활용하면 species가 인덱스가 아니라 컬럼으로 편입.
````

````python
iris_grouped_sum2 = iris.groupby(iris.species, as_index= False).sum()
iris_grouped_sum2
````

````python
iris_grouped_sum2[(iris_grouped_sum2.species=='setosa') | (iris_grouped_sum2.species == 'versicolor')]
# == iris_grouped_sum.loc[['setosa','versicolor']]
````

### pivot_table

````python
# 1) df.groupby(by = [그룹핑할열])[[함수를 적용해서 출력할 열들]].기초통계함수() 

# 2) 그룹화 방법 2
#    df.pivot_table(index = [그룹핑할열],
#                   values = [함수를 적용해서 출력할 열들],
#                    aggfunc = 함수(기본값:mean))
````

````python
display(iris.groupby(iris.species)[['petal_length','petal_width']].mean())
iris.pivot_table(index=[iris.species],
                values = ['petal_length','petal_width'],
                aggfunc='mean')
````

````python
from statsmodels.api import datasets
airquality = datasets.get_rdataset('airquality',
                                  package = 'datasets').data
airquality.head(2)
````

## 10.1 melt()를 이용한 언피벗팅(와이드포멧 => 롱포멧)

````python
airquality_melted = airquality.melt(id_vars = ['Month', 'Day'], 
                                    var_name='변수명', # default variable
                                   value_name='값') # default : value
````

````python
airquality_melted.sort_values(by = ['Month','Day'], inplace=True)
````

````python
airquality_melted.reset_index(drop = True)
````

````python
df = pd.read_csv(r'C:\ai_x\download\sharedata\부동산_250213\전국 평균 평당 분양가격(2013년 9월부터 2015년 8월까지).csv', encoding='cp949')
new_df = df.melt(id_vars=['지역'],var_name='연도월', value_name='평당분양가') # .sort_values('평당가격', ascending=False)
````

## 10.2 pivot_table()을 이용한 피벗팅 : 롱포멧 => 와이드포멧

````python
#    df.pivot_table(index = [그룹핑할열],
#                   values = [함수를 적용해서 출력할 열들],
#                   columns = ['기존 행에서 열로 묶을 기준'], 
#                    aggfunc = 함수(기본값:mean))

airquality2 = airquality_melted.pivot_table(index = ['Month','Day'],
                             columns = ['변수명'],
                             aggfunc='mean',
                             values = '값').reset_index()
airquality2.columns.name = None
airquality2
````

# 11절  데이터프레임(시리즈)에 함수 적용시키기
- df.apply(func, axis) : 각 축 기준 함수 적용(데이터프레임, 시리즈에서만 가능)
- df.applymap(func) : 각 요소 별 함수 적용(데이터프레임에서만 가능)
- map(func, iterable) : 각 요소별 함수 적용 (1차원 리스트, 시리즈에서만 가능)

## 11.1 apply

````python
member = pd.read_csv('data/ch11_member.csv', comment = '#')
member
````

````python
get_year = lambda datestr : int(datestr.split('-')[0])
get_month = lambda datestr : int(datestr.split('-')[1])
get_day = lambda datestr : int(datestr.split('-')[2])
print(get_year('1999-01-01'),get_month('1999-01-01'),get_day('1999-01-01'))
````

````python
member['year'] = member.Birth.apply(lambda datestr : int(datestr.split('-')[0]))
member['month'] = member.Birth.apply(lambda datestr : int(datestr.split('-')[1]))
member.drop(['Birth'], axis = 1)
````

````python
X = iris.iloc[:,:-2]
````

````python
# X의 열평균
avg = X.mean(axis = 0)
avg
````

````python
X.iloc[0] - avg
````

````python
# X의 모든 행에 대하여 열별 평균과의 거리를 소수점 2자리 까지 출력
X.apply(lambda x : round(x- avg,2), axis=1) # x는 1개 차원이라고 생각하고 설계
````

## ★ 연습 _ 2차원 데이터프레임으로 풀어보기

````python
# ★ 2차원데이터 프레임으로 풀어보기
def split_date(df) :
    df = df.copy() # 깊은 복사
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1][:-1])
    return df
````

````python
new_df.head(1)
new_df2 = new_df.apply(split_date, axis=1) # apply는 inplace가 없음
````

# 12절 일괄변경하기(결측치, 특정값)
## 12.1 fillna
- 대부분 결측치는 drop, 평균값 대체, 중위수 대체, 그룹별 평균값 대체, 그룹별 중위수 대체, 이전 값 대체, 이후 값 대체


````python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                 [3, 4, np.nan, 1],
                 [np.nan, 3, np.nan, 5],
                 [np.nan, 3, np.nan, 4],], columns = ['A','B','C','D'])
df
````

````python
# fillna : 모든 결측치를 대체
df.fillna(0)
````

````python
df.fillna(method='ffill') # ffill => forward fill 이전 값으로 대체
````

````python
df.fillna(method='bfill') # bfill => below fill 이후 값으로 대체
````

### 결측치를 열마다 특정 값으로 대체

````python
# 결측치를 열마다 특정 값으로 대체
values = {'A' : 85, "B":80, 'C' : 75, 'D':90}
df.fillna(values)
````

### 결측치를 열별 평균으로 대체

````python
avg = df.mean(axis=0)
avg
df.fillna(value = df.mean(axis=0)) # 중위수 등 기초 통계 수라면 mean 함수 자리만 바꿔주면 된다
````

````python
df.fillna(limit = 2, value = values) #limit 개의 결측치만 대체
````
