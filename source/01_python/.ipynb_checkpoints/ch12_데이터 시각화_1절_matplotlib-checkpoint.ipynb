## 1.3 그래프 영역 나누기
### 1) subplot() 함수로 서브플롯 추가

```python
import numpy as np
x = np.arange(0,10,0.01)
plt.figure(figsize=(8,6))
# subplot(전체화면 가로개수, 전체화면 세로개수, 위치 할당)
plt.subplot(2,1,1) ; plt.plot(x, np.sin(x))
plt.subplot(2,2,3) ; plt.plot(x, np.cos(x))
plt.subplot(2,2,4) ; plt.plot(x, np.sin(x)*np.cos(x)) # plt.subplot(224) 와 동일
plt.show()
```

```python
x = np.arange(0,10,0.01)
plt.figure(figsize=(8,6))
# subplot(전체화면 가로개수, 전체화면 세로개수, 위치 할당)
plt.subplot(2,2,1) ; plt.plot(x, np.sin(x))
plt.subplot(2,2,2) ; plt.plot(x, np.sin(x))
plt.subplot(2,2,3) ; plt.plot(x, np.cos(x))
plt.subplot(2,2,4) ; plt.plot(x, np.sin(x)*np.cos(x)) # plt.subplot(224) 와 동일
plt.show()
```

### 2) subplots() 함수로 서브플롯 추가
- 교안 14p

```python
# 2행 2열 서브플롯
x = np.arange(0,10,0.01)
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7,4)) # axes = nrows행 ncols열 크기의 행렬
axes[0,0].plot(x,np.sin(x), 'r:', linewidth = 1) ; axes[0,1].plot(x,np.cos(x), linestyle = '-')
axes[1,0].plot(x,np.tanh(x), 'g--') ; axes[1,1].plot([0,1,2,3,],[-1,-0.5,0,0.5],'k*-', linewidth = 0.5) 
axes[1,1].plot([0.5,1.5,2.5],[0,0.5,-1],'ro:')
axes[0,0].plot(np.random.randn(4,10),np.random.randn(4,10),'cs-.')
# linstyle = ':','--','-'
# marker = 'o','*','^','s'
plt.show()
```

```python
# 1행 2열
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (7,4)) # axes = nrows행 ncols열 크기의 행렬
axes[0].plot(x,np.sin(x), 'r:', linewidth = 1) ; axes[1].plot(x,np.cos(x), linestyle = '-')
```

## 1.4 다양한 그래프 그리기
- pyplot 함수들 https://matplotlib.org/stable/api/pyplot_summary.html
### 1) plt.scatter() 산점도 그래프

```python
plt.scatter(x=iris.petal_length,y=iris.petal_width,
           s=iris.petal_width*30, # size
           c = iris.petal_length*20, # color
           alpha = .8) # 투명도
```

### 2) boxplot(단점:분포를 알 수 없음) , violinplot()

```python
fig, axes = plt.subplots(ncols = 2)
axes[0].boxplot(iris.sepal_width)
axes[1].violinplot(iris.sepal_width)
axes[0].violinplot(iris.sepal_width)
plt.show()
```

### 3) bar(), barh(), axvline(수직선), axhline(수평선)

```python
x = [1,2,3]
y = [3,5,2]
fig, axes = plt.subplots(ncols = 3, figsize = (15,3))
axes[0].bar(x,y, color='limegreen',)
axes[1].barh(x,y, color = 'r') # x,y 좌표를 y,x로 사용하므로 plot을 추가할 때 x,y => y,x로 대입
axes[0].plot(x,y, 'ko-.')
axes[1].plot(y,x, 'ko-', linewidth = 2)
axes[2].axvline(0.5,color = 'g') # x=0.5 자리에 수직선 추가
axes[2].axhline(0.5,color = 'r', linestyle = '-.') # x=0.5 자리에 수직선 추가
```

### 4) hist(도수분포표;histogram)

```python
fig, axes = plt.subplots(nrows =2, ncols = 2, figsize = (10,6))
axes[0,0].hist(iris.sepal_length, bins = 10, color = 'r') # 히스토그램 ; 도수분포표
axes[0,1].hist(iris.sepal_length, bins = 15, cumulative = True) # 누적 히스토그램 ; 누적 도수분포표 => 총개수에 수렴 ★
axes[1,0].hist(iris.sepal_length, bins = 5, orientation = 'horizontal') # 수평 히스토그램
axes[1,1].hist(iris.sepal_length, bins = 10, histtype = 'step')
plt.show()
```

## 1.5 그래프 커스터마이징

### 1) linestyle(), linewidth(), color(), marker() ...... (plot의 매개변수로 지정)

```python
fig, axes = plt.subplots(nrows =2, ncols = 2, figsize = (10,6))
axes[0,0].hist(iris.sepal_length, bins = 10, color = 'r') # 히스토그램 ; 도수분포표
axes[0,1].hist(iris.sepal_length, bins = 15, cumulative = True) # 누적 히스토그램 ; 누적 도수분포표 => 총개수에 수렴 ★
axes[1,0].hist(iris.sepal_length, bins = 5, orientation = 'horizontal') # 수평 히스토그램
axes[1,1].hist(iris.sepal_length, bins = 10, histtype = 'step')
plt.show()
```

### 2) text(), annotate(), 

```python
x = np.linspace(0,10)
y = np.cos(x)
fig, axes = plt.subplots(nrows=1,ncols=2,figsize = (12,3))
axes[0].scatter(x,y, marker ='.')
axes[0].text(2,0,'Example Graphic', fontsize=20)
axes[1].scatter(x,y, marker ='*')
axes[1].annotate('COS', xy=(2,0), xytext=(4,0.5), arrowprops=dict(arrowstyle='->',connectionstyle='angle3'))
plt.show()
```
