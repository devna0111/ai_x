```python
import seaborn as sns

iris = sns.load_dataset('iris')
type(iris)

#1.
iris.head()

# 2.
x = iris.iloc[:,:-1]
y = iris.iloc[:,-1]
x, y

# 3. 
temp = iris.iloc[:50]
temp

# 4.
temp.describe()

# 5. 
iris.loc[iris.species == 'versicolor']
iris.loc[iris.species.str.contains('versicolor')]

# 6. 
iris_df = pd.concat((x,y),axis=1)
iris_df

# 7.
iris.mean(axis=0, numeric_only = True)

# 8.
iris.corr(numeric_only = True)

# 9.
iris.iloc[:,:-1].subtract(iris.mean(axis=0, numeric_only = True)).head()

# 10.
no10 = iris.groupby(iris.species).mean()
no10

# 11. 
no10

iris.head(1), iris.iloc[0]

def sub_mean(iris_series) :
    iris_series = iris_series.copy()
    iris_series - no10.loc[iris_series.species]
    return iris_series

sub_mean(iris.iloc[0])

iris.apply(sub_mean, axis=1).iloc[[0,1,2,50,51,52,100,101,102]]
```
